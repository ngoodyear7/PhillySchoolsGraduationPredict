```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
imp=SimpleImputer(missing_values=np.nan, strategy='mean',copy=False)
X = imp.fit_transform(X)
```

```python
X_train, X_test, y_train,y_test = train_test_split(X,y,train_size=0.66,test_size=0.34)
```

```python
regr = linear_model.LinearRegression()
```

```python
regr.fit(X_train, y_train)
```

```python
grad_y_predict = regr.predict(X_test)
```

```python
print('Coefficients: \n', regr.coef_)
print('Mean squared error: %.2f' % mean_squared_error(y_test, grad_y_predict))
print('Coefficient of determiniation: %.2f' % r2_score(y_test, grad_y_predict))
```
